```python
import requests
import numpy as np
from google.colab import autoviz
from matplotlib import pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import pandas as pd
import csv

response = requests.get("https://wuzzuf.net/search/jobs/?a=hpb%7Cspbl&q=product%20manager&start=0")
soup = BeautifulSoup(response.content, "html.parser")
type(soup)
```

```python
url = "https://wuzzuf.net/search/jobs/?a=hpb%7Cspbl&q=product%20manager&start="
page = 0
Final_Jobs = []

while True:
    active_url = url + str(page)
    response = requests.get(active_url)
    soup = BeautifulSoup(response.content, "html.parser")

    all_data = soup.find_all("div", {'class': "css-1gatmva e1v1l3u10"})

    if len(all_data) == 0:
      print("\nAll scrapping is done!")
      break
    for job in all_data:
      Title = job.find("div", {'class':"css-laomuu"}).find("a", {'class':"css-o171kl"}).text
      Link = "https://wuzzuf.net" + job.find("div", {'class': "css-laomuu"}).find("a", {'class': "css-o171kl"}).get("href")
      Company = job.find("div", {'class':"css-d7j1kk"}).find("a", {'class':"css-17s97q8"}).text.strip(" -")
      Location = job.find("div", {'class':"css-d7j1kk"}).find("span", {'class':"css-5wys0k"}).text
      Location = job.find("div", {'class':"css-d7j1kk"}).find("span", {'class':"css-5wys0k"}).text
      City = Location.split(", ")

      if City[-1].strip() == "Egypt":
        City = "Egypt"
      elif City[-1].strip() == "United Arab Emirates":
        City = "Emirates"
      elif City[-1].strip() == "Saudi Arabia":
        City = "Saudi"
      elif City[-1].strip() == "Kuwait":
        City = "Kuwait"
      elif City[-1].strip() == "Qatar":
        City = "Qatar"
      elif City[-1].strip() == "Libya":
        City = "Libya"
      elif City[-1].strip() == "Iraq":
        City = "Iraq"
      elif City[-1].strip() == "Uganda":
        City = "Uganda"
      elif City[-1].strip() == "Tunisia":
        City = "Tunisia"
      elif City[-1].strip() == "United States":
        City = "America"

      Duration_elem = job.find("div", {'class': "css-4c4ojb"})
      if Duration_elem is not None:
        Duration = Duration_elem.text
        if "days" in str(Duration):
          Is_Recent = True
        else:
          Is_Recent = False
      else:
        Duration = ""
        Is_Recent = False

      Skills = job.find("div", {'class':"css-y4udm8"})
      skills_text = Skills.text
      skills_lines = skills_text.split(" · ")

      Type = skills_lines[0]

      if "Full" in Type:
          Type = "Full Time"
      elif "Part" in Type:
          Type = "Part Time"
      elif "Work" in Type:
          Type = "Work from home"
      elif "Freelance" in Type:
          Type = "Freelance"
      elif "Shift based" in Type:
          Type = "Shift based"
      elif "Internship" in Type:
          Type = "Internship"
      else:
          Type = ""

      Experience_Type = skills_lines[0]

      if "Experienced" in Experience_Type:
          Experience_Type = "Experienced"
      elif "Manager" in Experience_Type:
          Experience_Type = "Manager"
      elif "Not specified" in Experience_Type:
          Experience_Type = "Not specified"
      elif "Senior Managment" in Experience_Type:
          Experience_Type = "Senior Managment"
      elif "Entry Level" in Experience_Type:
          Experience_Type = "Entry Level"
      elif "student" in Experience_Type:
          Experience_Type = "student"
      else:
          Experience_Type = ""

      Years_Of_Experience = skills_lines[1]

      if Years_Of_Experience[0].isdigit() == True:
        Years_Of_Experience = int(Years_Of_Experience[0])
      else:
        Years_Of_Experience = 0

      if Years_Of_Experience == 0:
          index = 1
      else:
          index = 2

      Skills_Set = skills_lines[index:]
      Skills_Count = len(Skills_Set)

      Final_Jobs.append({
          "Title": Title,
          "Link": Link,
          "Company": Company,
          "Location": Location,
          "City": City,
          "Is_Recent": Is_Recent,
          "Type": Type,
          "Experience_Type": Experience_Type,
          "Years_Of_Experience": Years_Of_Experience,
          "Skills_Count": Skills_Count,
          "Skills_Set": Skills_Set
          })
    if len(str(page)) == 1:
      print(f"Page number 0{page} is Scrapped")
    else:
      print(f"Page number {page} is Scrapped")
    page += 1
```

```python
df = pd.DataFrame(Final_Jobs)
df.to_csv("Product_Management.csv", index=False)
df
```

```python
import statistics

Years_Of_Experience_Values = [job["Years_Of_Experience"] for job in Final_Jobs]
Skills_Count_Values = [job["Skills_Count"] for job in Final_Jobs]

mean_years_experience = statistics.mean(Years_Of_Experience_Values)
stdev_years_experience = statistics.stdev(Years_Of_Experience_Values)

mean_skills_count = statistics.mean(Skills_Count_Values)
stdev_skills_count = statistics.stdev(Skills_Count_Values)

def is_outlier(value, mean, stdev, threshold=2):
    z_score = (value - mean) / stdev
    return abs(z_score) > threshold

filtered_years_experience = [value for value in Years_Of_Experience_Values if not is_outlier(value, mean_years_experience, stdev_years_experience)]
filtered_skills_count = [value for value in Skills_Count_Values if not is_outlier(value, mean_skills_count, stdev_skills_count)]

average_years_experience = statistics.mean(filtered_years_experience)
average_skills_count = statistics.mean(filtered_skills_count)

print(f"The average Years of experience (excluding outliers) is: {average_years_experience:.2f}")
print(f"The average skills count (excluding outliers) is: {average_skills_count:.2f}")
```

```python
def categorical_histogram(df, colname, figscale=1, mpl_palette_name='Dark2'):
  df.groupby(colname).size().plot(kind='barh', color=sns.palettes.mpl_palette(mpl_palette_name), figsize=(15*figscale, 4*figscale))
  plt.gca().spines[['top', 'right',]].set_visible(False)
  return autoviz.MplChart.from_current_mpl_state()

chart = categorical_histogram(df, *['Is_Recent'], **{})
chart
```

```python
def histogram(df, colname, num_bins=20, figscale=1):
  df[colname].plot(kind='hist', bins=num_bins, title=colname, figsize=(15*figscale, 4*figscale))
  plt.gca().spines[['top', 'right',]].set_visible(False)
  plt.tight_layout()
  return autoviz.MplChart.from_current_mpl_state()

chart = histogram(df, *['Skills_Count'], **{})
chart
```
